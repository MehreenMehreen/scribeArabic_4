```python
def get_special_ascii_set():
    char_set = []
    asccii_range = [[33, 64], [91, 96], [123, 126], 
                    # Arabic-Indic digits
                    [ord('\u0661'), ord('\u0669')]]
    for r in asccii_range:
        for val in range(r[0], r[1]+1):
            char_set.append(chr(val))
    return char_set

print((ord('\u0661')))
special_ascii = get_special_ascii_set()
special_ascii.sort()
print(special_ascii)
special_ascii = set(special_ascii)
```

```python
larger_set = set(larger_set)
smaller_set = set(smaller_set)
l_diff_ss = larger_set.difference(special_ascii.union(smaller_set))
l_diff_ss = (list(l_diff_ss))
l_diff_ss.sort()
for l in l_diff_ss:
    print(l, ascii(l))
```
